## 6. Remove duplicates

now lets try to remove duplicates. But when doing so we first need to identify a primary key or a key as to which we can define duplicatates. A key would be such that it is the shortest mixture of variables (maybe even only one) to identify a unique and distinct job from another in the dataset. 

- We know that the title is unique. However according to me the title cannot be a good key for such a database. This database consists of only jobs for about a couple of years. with time there might be a job with the same title in future which would not be the same as the one which was posted multiple years ago with the same title. Each job posting is dictinct. 
    - For example. An analyst position is open today. 10 years later that analyst is fired and the same company with the same location, category etc. posts the vacancy for the job for the same title. Hence the title would not remain the primary key to distinguish between the two similar jobs even though they are 10 years apart. 
- The id is unique howeve;r, the ID is a surragate key and does not itself represent different jobs. The datasets may be from different databases and even though they are both 8 digits long with integers the function that creates them may be different. Hence even though the ids are different in the different datasets the jobs might be the same. 
- Hence the key should be such that each job is different. In my opinion there should be a composite key (a key with a mixtrue of attributes). In my opinion to identify a unique and distinct job it should have the following variables in the key:
    - Title: title should be the same if it s the same job.
    - Location: the same title can be used for differnt branches of the company hence to distinguish between them location is included
    - ContractType: The type of the contract. 2 business analyst might be getting hired by the company however, one can be partime, and second full time, even though the location and title are the same.
    - ContractTime: Same exampel as contract type but the timings can be different one might be permanetn other might be contract
    - Company: The company needs to be the same for the job to be the same. 2 companies can hire business analysts full time permanent but they are different jobs.
    - Category: The cateory of the job should be the same. For example an analyst could be in the Finacne cateogry or some other. 
    - Salary per annum: The salary posted for the job needs to be the same for it to be considered a correct job.
    - OpenDate: The time for the job should match. The same job may be posted in the future if the vacancy occurs
    - CloseDate: Closing data is important as two positions might be open but one might need to be hired before the other. 
    - Source name: Depends, but in this case will be included. 
    
The source name may or may not be included as a company might post for the job themselves or might also give details to an agency to post aswell. However it depends, if the database is only to keep one row for each job the source name should not be included however if the database wants to store all the job postings and considers them different then source name should be included in the key. 


In this case, we consider the jobs posted by differnt sources as different. Choosing the key is always a task that is discussed with the client and decided based on the business rules. However since, the client is not present, it is alwasy best to be on teh safe side. As once the primary key is thought of the integrity constraints set in the incase the key needs to be changed in the future it causes alot of issues due to schema change of the database as well as the dashboards, or apps connected to the database that use that key to obtain data from the database. 

Hence lets see what the chances of the duplicates are for id, Title and with the above composite key mentioned. 

```python
#only id
title_cls =['Id']
sum(df12.duplicated(subset=title_cls))
```

```python
#only title
title_cls =['Title']
sum(df12.duplicated(subset=title_cls))
```

```python
#all#composite pk
all_cls =['Location','Category', 'Salary per annum', 'SourceName', 'OpenDate', 'CloseDate','ContractTime', 'Company', 'Title', 'ContractType']
sum(df12.duplicated(subset=all_cls))
```

It seems that the count of duplicates with the title or with the other total composite keys seem to be the same. However we will still remove the duplicates based on the composite primary key. But before we do so lets see if the 400 duplicates for both are the same duplicates or different ones. 

```python
#creating set of ids for title duplicates as well as duplicates for all attributes
id_title_set = set(df12.loc[df12.duplicated(subset=title_cls, keep=False),'Id'].tolist())
id_all_set = set(df12.loc[df12.duplicated(subset=all_cls, keep=False),'Id'].tolist())

#checking whethere there are values in the set atre not common in both of them 
id_title_set ^ id_all_set
```

We can see that the two keys represnt the same dataset duplicates, hence lets remove the duplicates.

```python
len(df12)
```

```python
merged_df = df12.drop_duplicates(subset=all_cls)
```

```python
len(merged_df)
```
