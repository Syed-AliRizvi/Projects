```python
excel_data = pd.ExcelFile('basic_indicators.xlsx')
excel_data
```

```python
#check sheet names
excel_data.sheet_names
```

```python
#parse excel file into dataframe 
df = excel_data.parse('Basic Indicators')
```

```python
#check the dataframe
df
```

```python
#change column names for easy access
df.columns = range(df.shape[1])
```

```python
#drop unwanted cols 
df.drop([0] + [15] + list(range(17,24)), axis=1, inplace=True)
```

```python
#since we cant see the end of our wanted data in the datafrane lets search the end of our table by looking for zimbabwe
df[df.iloc[:,0] == 'Zimbabwe']
```

```python
#now that we know the index check the dataframe around that record to check where to drop rows from
df.iloc[200:210,:]
```

```python
#now that we know the row indices that we dont want we can sinmply drop them 
df.drop(list(range(0,4)) + list(range(206,df.shape[0])), axis=0, inplace=True)
```

```python
#reset index change index name as well as col names
df.set_index(keys=1, inplace=True)
df.index.name = 'Country Name'
df.columns = range(df.shape[1])
```

```python
#view the corrected dataframe
df
```

```python
#now that the data is in the correct form we check the data types and fix them incase they are not numeric
df.dtypes
```

```python
#convert data into numeric where ever possible
df = df.convert_objects(convert_numeric=True) #convert cols to numeric
```

```python
# check updated data types
df.dtypes
```

```python
#since data is no correct we check the dataframe one final time.
df
```

```python
#now we transer it into a csv file
#before we do so the csv file required needs only integers while we have float hence we will do this while converting to csv
df.to_csv('basic_indicators.csv', float_format='%.0f')
```
